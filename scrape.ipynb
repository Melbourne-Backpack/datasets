```python
url = 'https://www.zillow.com/melbourne-fl/rentals'
```

```python
html = requests.get(url=url,headers=header)
html.status_code
```

```python
bsobj = soup(html.content,'lxml')
```

```python
price_list = []

for price in bsobj.findAll('div',{'class':'list-card-heading'}):
    price_list.append(price.text.replace('bd','b|').replace('|s','|').replace('o','o|').strip().split('|')[:-1])
price_list
```

```python
image = []

for price in bsobj.findAll('div',{'class':'list-card-top'}):
    price_list.append(price.text.replace('bd','b|').replace('|s','|').replace('o','o|').strip().split('|')[:-1])
image
```

```python
address = []

for adr in bsobj.findAll('div',{'class':'list-card-info'}):
    address.append(adr.a)

address
```

```python
df = pd.DataFrame(price_list,columns=['Price1','Price2','Price3', 'Price4'])
df
```

```python
df['Address'] = address
df
```

```python
df.to_csv('data/zillow.csv')
```

# DWELL STUDENT LIVING NEAR RMIT UNIVERSITY (15 mins walk)

```python
from bs4 import BeautifulSoup
import requests
from csv import writer

url= "https://www.dwellstudent.com.au/melbourne/#rooms"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
lists = soup.find_all('div', class_="col-md-6 item")

with open('housing.csv', 'w', encoding='utf8', newline='') as f:
    thewriter = writer(f)
    header = ['Title', 'Type', 'Price', 'Image']
    thewriter.writerow(header)

    for list in lists:
        title = list.find('div', class_="title")
        type = list.find('div', class_="icon-number") #living, bed, bath numbers
        price = list.find('div', class_="price-wrapper")
        image = list.find('div', class_="img bg-cover-center")
        
        info = [title, type, price, image]
        thewriter.writerow(info)
```

```python
print(info)
```

```python
df = pd.read_csv('housing.csv')
df
```

```python
print(f'NUMBER OF CATEGORIES: {df.Image.nunique()}; \n\nUNIQUE NAMES OF THE CATEGORIES {df.Image.unique()}\n\n\n')
```

```python
df['Title'] = df['Title'].str.replace(r'<[^<>]*>', '', regex=True)
df['Type'] = df['Type'].str.replace(r'<[^<>]*>', '', regex=True)
df['Price'] = df['Price'].str.replace(r'<[^<>]*>', '', regex=True)
# df['Image'] = df['Image'].str.replace(r'<[^<>]*>', '', regex=True)

import re
exp = re.compile(r'(?P<url>https?://[^\s]+)')
df[['Image']] = df['Image'].str.extract(exp, expand=True)

df.to_csv('data/dwell.csv')
df
```
